#**Import Library**

```python
import matplotlib.pyplot as plt
import matplotlib
from matplotlib.pyplot import figure
import numpy as np
import pandas as pd
```

# **Scatter Plot**

```python
titanic = pd.read_csv('https://raw.githubusercontent.com/zhafar3adib/belajar_datascience/master/dataset/train.csv')
titanic.head()
```

```python
titanic2 = pd.pivot_table(titanic, values='Fare',index='Age',aggfunc='sum')
titanic2 = titanic2.reset_index()
titanic2
```

```python
colors=titanic2['Fare']
```

```python
plt.scatter(
    x=titanic2['Age'], #data untuk x axis
    y=titanic2['Fare'], #data untuk y axis
    s = 100, #ukuran marker
    c =colors, #gradasi warna berdasarkan variabel
    cmap = 'RdYlGn', #jenis warna
    marker = '^', #bentuk marker
    edgecolor = 'lightblue', #warna pinggiran marker
    linewidths = 1, #ukuran pinggiran marker
    alpha = 1 #1 = opaque, 0 = transparent
)
plt.colorbar()
plt.show()
```

# **Line Chart**

      > encoding="ISO-8859-1" ,used to avoid errors "'utf-8' codec can't decode".


```python
#reading csv dataset
data = pd.read_csv('https://raw.githubusercontent.com/nullbytee/DigitalSkola_DataScience_Batch11/main/dataset/Sample_Superstore.csv', encoding="ISO-8859-1")
data.head()
```

```python
#data subset
data_line = data[['Order Date','Sales','Profit']].sort_values('Order Date')

#converting into DateTime
data_line['Order Date'] = pd.to_datetime(data_line['Order Date'])

#groupby to get the average Sales and Profit on each day
data_line = data_line.groupby('Order Date').mean()

#showing results
data_line.head()
```

        > Use 'figsize=[x,x]' to customize the size of the figure.
        > Use 'color='#xxxxx' to add custom color of the line.
        > Use 'size=xx' to set font size in the title.

```python
#visualizing the Line Chart figure size
plt.figure(figsize=[12,5])

#avg sales overtime
plt.plot(data_line.index, 'Sales', data=data_line, color='#F05454')

#avg profit overtime
plt.plot(data_line.index, 'Profit', data=data_line, color='#30475E')

#set title and font size
plt.title("Average Sales and Profit over Time Period (2014-2018)", size=12)

#showing results
plt.show()
```

# **Bar Plot**

```python
consolegame = pd.read_csv('https://raw.githubusercontent.com/Indra2101/digitalskola_ds11/master/Video_Games_Sales_as_at_22_Dec_2016.csv')
consolegame.head()
```

```python
consolegame2 = pd.pivot_table(consolegame, values='Global_Sales',index='Platform',aggfunc='sum')
consolegame2 = consolegame2.reset_index()
consolegame2.head()
```

```python
#Script for Colored Vertical Bar Plot
x = consolegame2['Platform']#Variabel untuk Axis X
y = consolegame2['Global_Sales']#Variabel untuk Axis Y

f, ax = plt.subplots(figsize=(20,5))#mengatur lebar dan tinggi bar (width,height)
plt.bar(x, y, color = "cyan") #mengatur warna bar chart
plt.show()
```

```python
#Script for Colored Horizontal Bar Plot
x = consolegame2['Platform']#Variabel untuk Axis X
y = consolegame2['Global_Sales']#Variabel untuk Axis Y

f, ax = plt.subplots(figsize=(5,20))#mengatur lebar dan tinggi bar (width,height)
plt.barh(x, y, color = "pink") #mengatur warna bar chart
plt.show()
```
